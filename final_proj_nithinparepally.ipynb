<center><h1>Introduction</h1></center>

Alzheimer's disease, the most common form of dimentia, is a neurodegenerative disease that affects around 6.9 million Americans as of 2023. It is a progressive disease, mainly affecting ages 65 or above and less common in younger people that starts with minimal memory loss and gradually results in the loss of ability in carrying out daily activities. According to the CDC, it is projected that around 14 million Americans will have Alzheimer's Disease by 2060. The disease targets parts of the brain that control memory, coginitive functions, language and judgement. Apart from progressive nature of the disease and what it entails for the (essentially considering the severity of impact the disease has on an individual's well being), another reason for why scientists are interested in learning more about the disease is that death rates for Alzheimer's Disease have been increasing over time, unlike death rates of cancer or heart disease [(CDC)](https://www.cdc.gov/aging/alzheimers-disease-dementia/about-alzheimers.html#:~:text=Death%20rates%20for%20Alzheimer%27s%20disease,that%20are%20on%20the%20decline.&text=Dementia%2C%20including%20Alzheimer%27s%20disease%2C%20has,Alzheimer%27s%20may%20be%20considerably%20higher.). 

##### Subjective Cognitive Decline
The analyses that will be conducted in this tutorial will be on the basis of Subjective Cognitive Decline. Subjective cognitive decline is known to be one of the [earliest noticeable symptoms of dementia and Alzheimer's disease](https://www.cdc.gov/aging/data/subjective-cognitive-decline-brief.html#:~:text=Subjective%20Cognitive%20Decline%20(SCD)%20is,Alzheimer%27s%20disease%20and%20related%20dementias.). It is based on an individual's perception of cognitive changes, impairment, and loss of memory. The symptom is important due to its early detection nature. Individuals who suspect their cognitive decline to be worsening can seek professional help and potentially have interventions be done in the form of lifestyle changes or treatments to the slow the development of any potential dementia related diseases. Some common examples of Subjective Cognitive Decline include forgetting to perform daily or routine tasks, failing to make basic judgements, misplacing objects and more.

In the data processing stage of the data science pipeline, we will be preparing the data collected from the previous stage to fit the needs of specific data explorations conducted in the Exploratory Data & Analysis stage and the analyses conducted in the Analysis and Hypothesis Testing stage. Once our data is processed or "cleaned", we have what is called a tidy dataset. To learn more about tidy datasets, check out this article: [Tidy Data](https://www.jstatsoft.org/article/view/v059i10)

We will first remove any uneccessary rows and columns of the two datasets.

```python
# remove any unecessary columns and columns that have missing values for every observation
rmv_cols = ['Datasource', 'StratificationCategory3', 'Stratification3',
            'StratificationCategoryID3', 'StratificationID3', 'Report', 'ResponseID',
            'DataValueTypeID', 'Response', 'Sample_Size', 'Data_Value_Unit', 'StratificationCategory1', 
            'Data_Value_Footnote', 'Data_Value_Footnote_Symbol']

df.drop(rmv_cols, inplace=True, axis=1)
death_df.drop('URL', inplace=True, axis=1)

# remove any rows from the death_df dataset that correspond to years before 2015
death_df = death_df[death_df['YEAR'] >= 2015]

display(df.head())
display(death_df.head())
```

The stratification columns of the healthy aging data actually divide the data based on three metrics, age group, gender and ethnicity. The following code ensures that these metrics have their own appropriately named columns in the healthy aging data. The Data_Value_Type column specifies the specific value each row reports which include Mean and Percentage. We will also turn these two variables into their own distinct columns for convenience in the Exploratory Data Analysis and Data visualization stage of the data science pipeline.

```python
# rename the Stratification1 columns to Age Group
df = df.rename(columns = {"Stratification1":"Age Group"})

# column variables are stored in the column Stratification2 so we will turn these variables into their
# own distinct columns
strat2_pivot_df = df.pivot(columns='StratificationCategory2', values='Stratification2')
strat2_pivot_df = strat2_pivot_df[['Gender', 'Race/Ethnicity']]
df = pd.concat([df, strat2_pivot_df], axis=1)
df.drop(['StratificationCategory2', 'Stratification2'], inplace=True, axis=1)

# column variables are also stored in the column Date_Value_Type so we will turn these variables into their
# own distinct columns
data_value_pivot_df = df.pivot(columns='Data_Value_Type', values='Data_Value')
df = pd.concat([df, data_value_pivot_df], axis=1)

# reorder the columns in the healthy aging data : we want ID columns last
df = df[['RowId', 'YearStart', 'YearEnd', 'LocationAbbr', 'LocationDesc', 'Class',
       'Topic', 'Question', 'Mean', 'Percentage', 'Data_Value_Alt',
       'Low_Confidence_Limit', 'High_Confidence_Limit',
       'Age Group','Gender', 'Geolocation',
       'LocationID', 'StratificationCategoryID1', 'StratificationID1',
       'StratificationCategoryID2', 'StratificationID2']]

df.head()
```

Now we will create and store data frames corresponding to different topics including Caregiving, Mental Health, Nutrition/Physical Activity/Obesity, Overall Health, Screenings and Vaccines, and Smoking and Alcohol Use. By doing this, we will be able to classify potential risk factors of Alzheimer's disease so that in the exploratory data analysis stage, we can explore trends and relationships corresponding to different areas of health and living.

```python
# make a copies of the Alzheimer's Mortality data frame and the Healthy aging data
source_healthy_df = df.copy()
source_death_df = death_df.copy()

# remove commas from data values and convert the values to appropriate types
source_death_df['DEATHS'] = source_death_df['DEATHS'].str.replace(',', '').astype(int)
source_death_df['RATE'] = death_df['RATE'].astype(float)

# create seperate data frames grouped by class
grouped = df.groupby(by='Class')
dfs_groupedby_class = {class_name: group for class_name,group in grouped}

# create a function that returns a dictionary where data frames grouped by topic can be retrieved by their label
def create_dfs_groupby_topic(class_name):
    
    # retrieve and store a class dataframe with the name class_name
    temp_df = dfs_groupedby_class[class_name]
    
    # create groupby objects based on topic
    grouped_topic = temp_df.groupby(by='Topic')
    
    # create a dictionary storing topic names as keys and dataframes grouped by the respective topic names as values
    dfs_grouped_topic = {name: group for name,group in grouped_topic}
    
    return dfs_grouped_topic

class_dfs = {}
df_names = {
    'caregiving_df': 'Caregiving', 
    'm_health_df': 'Mental Health', 
    'nutrition_df': 'Nutrition/Physical Activity/Obesity', 
    'overall_health_df': 'Overall Health', 
    'screenings_and_vaccines_df': 'Screenings and Vaccines', 
    'smoke_and_alcohol_df': 'Smoking and Alcohol Use',
    'cog_dec_df':'Cognitive Decline'
}

# create dataframes for all topics in the healthy aging data
for df_name, class_name in df_names.items():
    class_dfs[df_name] = create_dfs_groupby_topic(class_name)

# helper function that returns dataframes corresponding to the class with name df_name
def get_dfs(df_name):
    return class_dfs[df_name]

# helper function that returns a list of all the topics under a specific class
def get_topics(dfs):
    return [topic for topic in dfs]
```

The data in the healthy aging dataset is divided on location relating to states and territories in the United states as well as regions. Each region, state and territory has its own ID. We need to store these Ids somewhere so that we can filter by regions or states and territories in the future. Depending on our use case, we may want to look at data corresponding specifically to either regions or states.

```python
# get region, states, and territory IDs
locID_locDESC_locABBR = {}
for index, row in source_healthy_df.iterrows():
    locID = row['LocationID']
    locDESC = row['LocationDesc']
    locABBR = row['LocationAbbr']
    
    locID_locDESC_locABBR[locID] = locABBR, locDESC
print(locID_locDESC_locABBR)
```

To help conduct explorations of the healthy aging data in the exploratory data analysis and data visualization stage of the data science pipeline, the following code details some helper filter functions that will allow us to filter our data based on different metrics including location, overall age, different age groups, gender, and overall population.

```python
united_states_ID = 'US'
region_IDS = ['WEST', 'NRE', 'MDW', 'SOU']
states_and_territories_IDS = [abbr for key, (abbr, desc) in locID_locDESC_locABBR.items() 
                              if abbr != united_states_ID and abbr not in region_IDS]
# function that filters the given dataframe by the given location
# which can be the entire United States (and it's territories), region or state.
def filter_location(df, location):
    if location == "states_and_territories":
        return df[df['LocationAbbr'].isin(states_and_territories_IDS)]
    elif location == "regions":
        return df[df['LocationAbbr'].isin(region_IDS)]
    elif location == "us":
        return df[df['LocationAbbr'] == united_states_ID]

# filters given dataframe by overall age
def filter_overall_age(df):
    return df[df['Age Group'] == 'Overall']

# filters given dataframe by different age groups
def filter_age_groups(df):
    return df[df['Age Group'].isin(['65 years or older', '50-64 years'])]

# filters given dataframe by gender
def filter_gender(df):
    return df[df['Gender'].isin(['Female', 'Male'])]

# filters given dataframe by the overall population (in the current dataframe)
# this project mainly focuses on drawing insights with respect to age and gender
def filter_ethnicity_overall(df):
    return df[df['StratificationID2'] == 'OVERALL']

# filters given dataframe by overall population (in age, ethnicity, and gender)
def filter_overall(df):
    temp_df = df.copy()
    temp_df = temp_df[temp_df['StratificationID1'] == 'AGE_OVERALL']
    temp_df = temp_df[temp_df['StratificationID2'] == 'OVERALL']
    return temp_df
```

In this section of the data science pipeline, we will look for some trends or potential relationships between different healthy aging factors, Alzheimer's Disease mortality, and symptom reports (concerning cognitive decline), with the help of some data visualization and analysis tools including libraries like MatplotLib, Seaborn, Plotly, Sklearn and more. From this stage forward, because the healthy aging dataset used for this project is missing information relating to some states and territories entirely, please keep in mind that the insights and generalizations that will be made may or may not truly represent the real world trends and behaviors.

Lets start of by analyzing the Alzheimer Mortality data. We will first look at how the number of deaths caused by Alzheimer's disease varies across most US states over the period from 2015 to 2021. This will give us a starting point in uncovering potential relationships and correlations between healthy aging factors and Alzheimer's Disease Mortality.

```python
import matplotlib.pyplot as plt
import seaborn as sb

# create a dataframe that counts the number of deaths across the years 2015-2021 for each state
plt_death_df = (
    source_death_df[['STATE', 'DEATHS']]
    .groupby('STATE')
    .sum()
    .sort_values(by='DEATHS', ascending=False)
)

# define plot size
plt.figure(figsize=(35,15))

# make a bar plot with the data stored in plt_death_df
sb.barplot(x=plt_death_df.index, y='DEATHS', data=plt_death_df, order=plt_death_df.index)

# provide appropriate titles and labels
plt.title("Number of Deaths caused by Alzheimer's Disease across the years 2015-2021 by State")
plt.xlabel('States')
plt.ylabel('Number of Deaths across the years 2015-2021')
```

From this bar plot alone, we see that California ranks first in total number deaths caused by Alzheimer's disease relative to other states, followed by Texas, Florida, and Ohio. However, the bar plot fails to account for the populations of the different states. California and Texas for instance are generally more populous than Vermont or Maryland. Let try and plot death rates instead (number of deaths per 100,000 total population) and see if the same observations hold (i.e California ranking first again). By using the death rate, we will be accounting for state population which provides for a more suitable metric for comparing Alzheimer mortality between different states. 

```python
grouped_years = death_df[['YEAR', 'STATE', 'RATE']].groupby('YEAR')

for (year, df) in grouped_years:
    
    df = df.sort_values(by='RATE', ascending=False)
    df = df[['STATE', 'RATE']]
    
    plt.figure(figsize=(30, 5))
    sb.barplot(x='STATE', y= 'RATE', data=df, order=df['STATE'])
    plt.title(f"Alzheimer's Disease Death Rate by State in {year}")
    plt.xlabel('States')
    plt.ylabel('Death Rate')
    
    # give a rotation to the ticks corresponding to different state ids so that they appear neatly
    plt.xticks(rotation=45)
    
    plt.show()
```

From the bar plots above we no longer see California ranking first. According to the [CDC's Alzheimer's Disease page](https://www.cdc.gov/aging/publications/features/alzheimers-disease-deaths.html) people should see the doctor if they "experience symptoms such as memory loss affecting their daily life, difficulties with problem solving, or misplacing objects". In the healthy aging data, under the Cognitive Decline Class category, people from the United States, DC, and territories were asked questions in relation to: <br><br>
<div style='text-align:left'>
    • Functional difficulties associated with subjective cognitive decline or memory loss among older adults <br>
    • Need assistance with day-to-day activities because of subjective cognitive decline or memory loss <br>
    • Subjective cognitive decline or memory loss among older adults <br>
</div>

Results corresponding to the responses to these questions were recorded a percentages per state and territory (under the United States). Based on what we observed in the bar and line plots above, we should generally expect an overall linear growth in the percentages of people experiencing functional difficulties as a result of cognitive decline among older adults, needing assistance with day to day activities, and reports of subjective cognitive decline among older adults. If we do indeed observe this, then it essentially validates these categories as a metric we could compare other potential risk factors to. The following code generates plots which display these percentages across most states and territories on the y axis and year (2015-2021) on the x axis.

```python
# helper function that filters the given dataframe by location corresponding to the entire united states
# and its territories, overall age, and overall population (not split by gender or ethnicity)
def shape_cog_dec_overall(df):
    plt_df = df.copy()
    
    # filter the given dataframe by the entire united states and its territories and by 
    # overall age and overall population (not split by gender or ethnicity)
    plt_df = filter_location(plt_df, 'us')
    plt_df = filter_overall(plt_df)
    
    # filter out rows where the start and end years are not the same
    plt_df = plt_df.sort_values(by='YearEnd', ascending=True)
    plt_df['YearEnd'] = plt_df['YearEnd'].astype(int)
    plt_df['YearStart'] = plt_df['YearStart'].astype(int)
    plt_df = plt_df[(plt_df['YearEnd'] - plt_df['YearStart']) == 0]
    
    return plt_df

plt_func_diff_cog_dec_df = shape_cog_dec_overall(func_diff_cog_dec_df)
plt_need_assistance_cog_dec_df = shape_cog_dec_overall(need_assistance_cog_dec_df)
plt_subjective_cog_dec_df = shape_cog_dec_overall(subjective_cog_dec_df)
plt_talked_abt_subjective_cog_dec_df = shape_cog_dec_overall(talked_abt_subjective_cog_dec_df)

dfs = [
    plt_func_diff_cog_dec_df, 
    plt_need_assistance_cog_dec_df, 
    plt_subjective_cog_dec_df, 
    plt_talked_abt_subjective_cog_dec_df
]

colors = ['red', 'green', 'blue', 'purple']
lbls = [
    'report cognitive decline interferes with their ability to engage in social activities or household chores',
    'report needing assistance in daily activities due to cognitive decline',
    'report subjective cognitive decline happening more often or getting worse (preceding 12 months)',
    'report talking with a health care professional about their subjective cognitive decline'
]

# make a scatter plot with regression lines corresponding to the four topics in the data corresponding to
# the cognitive decline class
plt.figure(figsize=(20,5))

for i, df in enumerate(dfs):
    X = df['YearStart']
    y = df['Percentage']
    
    plt.scatter(X, y, color=colors[i], label=lbls[i])
    
    m, b = np.polyfit(X, y, 1)
    plt.plot(X, m*X+b, color=colors[i])
    
    # print the correlation coefficient of each regression line
    print(lbls[i], '\n • correlation coefficient', X.corr(y), '\n')
    
plt.title('Percentages of older adults reporting different behaviors relating to cognitive decline')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.legend()
plt.show()
```

From the scatter plot above and the respective regression lines, we mostly see a sideways trend in the behaviors and effects exhibited by older adults as a result of cognitive decline. The correlation coefficients of the respective regression lines indicate weak negative correlations between year and percentage of older adults reporting how cognitive decline is interfering with their ability to engage in social activities or household chores, needing assistance in daily activities, and talking about their subjective cognitive decline to a health care professional, and a weak positive correlation between year and percentage of older adults reporting cognitive decline happening more often or getting worse in the preceding 12 months.

By looking at the age group of overall ages and the resulting weak correlations, we cannot conclusively claim that there exists some distinct trend in the behaviors and effects exhibited by older adults as a result of cognitive decline over time. So, lets try and see if there is a trend amongst a specific age group. The CDC claims that age is the best known risk factor for Alzheimer's disease and cognitive decline. Based on this, we might expect to see a distinct trend in the age group 65 years or older. Let us see if that is indeed the case.

### Preparing the data for analysis
Our first analysis of the healthy aging data will consist of running an ordinary least squares regression on the healthy aging data filtered by states and territories and the overall age group. As aforementioned, we will be treating the factors from the cognitive decline class as our dependent variables. We will be using these factors as a metric to compare the predictive power of the factors from the other classes. The main goal of running an ordinary least squares regression is to minimize the sum of squared differences between the actual and predicted values. In doing so, our result will consist of the best estimates for the coefficients of the line of best fit. The coefficients can either have a positive or negative sign. A positive sign indicates that it's relationship with the dependent variable is one where as the independent variable increases then the dependent variable also tends to increase. Conversely, a negative sign indicates that it's relationship with the dependent variable is one where as the independent variable increases then the dependent variable tends to decrease (vice versa). This will become important later on. But before we move forward, we need to process our data even further. To best run the ordinary least squares regression, we would need all of the values for each of the factors from all of the classes in their own seperate columns with header names titled appropriately (topic name).

To get the best results from the oridnary least squares regression, we will be filtering the healthy aging data by state and territory. Because of this, we see that there exists some missing values. Missing values is something that is commonly observed when preparing data. They can be missing due to a wide variety of reasons and are classfied primarily into three categories: Missing Completely at Random, Missing At Random, and Missing Not at Random. Missing Completely at Random means that the missiningness of the data does not depend on the obeserved or unobserved data. Missing at Random means that the missiningness of the data is dependent on the observed data. This would mean that there exists some mechanism that we could model to account for the missingness. Missing not at random means that the missingness of the data has to do with the missing data itself. For instance, the nature of questions asked could be the cause of why the data is missing where people might be hesitent to answer. Common ways to deal with missing values include Mean Imputation, Hot deck imputation, Cold deck imputation and forms of Multiple Imputation. For this tutorial, we will be using Mean Imputation and Hot deck imputation. In Mean Imputation, we replace missing values with the average of all the observed (not missing) values. In Hot deck imputation we replace missing values with observed values whose subject is most like the subject who's value is missing.

To learn more about imputation, check out these articles and resources: <br>

### Oridnary Least Squares Regression Model

```python
from sklearn.impute import KNNImputer

# same filter function from the EDA stage will be used to filter the data in this stage
def shape_cog_dec_overall(df, loc):
    plt_df = df.copy()
    
    # filter the given dataframe by the entire united states and its territories and by 
    # overall age and overall population (not split by gender or ethnicity)
    plt_df = filter_location(plt_df, loc)
    plt_df = filter_overall(plt_df)
    
    # filter out rows where the start and end years are not the same
    plt_df = plt_df.sort_values(by=['YearEnd', 'LocationAbbr'], ascending=True)
    plt_df['YearEnd'] = plt_df['YearEnd'].astype(int)
    plt_df['YearStart'] = plt_df['YearStart'].astype(int)
    plt_df = plt_df[(plt_df['YearEnd'] - plt_df['YearStart']) == 0]
    
    return plt_df

import re

def create_combined_df(loc):
    # retrieve IDS based on filter location
    if loc == 'states_and_territories':
        IDS = states_and_territories_IDS
    elif loc == 'regions':
        IDS = region_IDS
        
    years = []
    states = []
    
    for year in [2015, 2016, 2017, 2018, 2019, 2020, 2021]:
        for location in IDS:
            years.append(year)
            states.append(location)

    # intialize the combined data frame that will be returned
    init_data = {'YearStart': years, 'YearEnd': years, 'LocationAbbr': states}
    combined_df = pd.DataFrame(init_data)
    combined_df = combined_df.sort_values(by=['YearEnd', 'LocationAbbr'])

    def pivot_topic_value(df):

        columns = ['YearStart', 'YearEnd', 'LocationAbbr', 'Topic', 'Mean', 'Percentage']
        temp_df = df[columns].copy()
        temp_df = temp_df.reset_index()[columns]

        mean_or_percentage = 'Mean' if temp_df['Mean'].notna().any() else 'Percentage'
        
        # remove spaces in the topic and create a new column with modified topic header
        topic = re.sub('[^A-Za-z0-9]', '_',('_'.join(temp_df['Topic'].unique()[0].split()))).lower()
        temp_df[topic] = temp_df[mean_or_percentage].astype(float)
        temp_df = temp_df[['YearStart', 'YearEnd', 'LocationAbbr', topic]]

        combined_df[topic] = temp_df[topic]

        # perform imputation on missing values
        if combined_df[topic].isna().any():
            
            # calculate the mean by year
            mean = temp_df.groupby('YearStart').mean()[topic]
            
            for index, row in combined_df.iterrows():
                year = row['YearStart']
                if pd.isna(row[topic]):
                    
                    # if the year is missing, replace missing values with the closest value (using KNNImputer)
                    if year not in mean.index:
                        temp = mean.copy()
                        imputer = KNNImputer(n_neighbors=2)
                        temp[year] = np.nan
                        impute_vals = imputer.fit_transform(temp.values.reshape((-1,1)))
                        impute_vals = list(zip(temp.index, impute_vals))
                        get_impute_val = lambda x: next(j[0] for i,j in impute_vals if x == i)
                        combined_df.at[index, topic] = get_impute_val(year)
                    else:
                        # replace missing values by the mean 
                        combined_df.at[index, topic] = mean.loc[year]

    # iterate through all classes and pivot topic names and values into distinct columns
    for name in df_names:
        dfs = get_dfs(name)
        topics = get_topics(dfs)
        for topic in topics:
            df = shape_cog_dec_overall(dfs[topic], loc)
            pivot_topic_value(df)
            
    return combined_df

combined_df = create_combined_df('states_and_territories')
combined_df.head()
```

Now that we have cleaned the data, we will will run an ordinary least squares regression using the ols function from the statsmodels.formula.api import. Because we mentioned that we have three dependent variables from the cognitive decline class, in order see the effect of the factors from the other classes, we will include an interaction term. A term that captures the dependence relation between the independent variables and the factors we identified from the cognitive decline class. After running the interaction ordinary least squares regression, we will print statistical information that details the values of each coefficient, the p values of each coefficient (values that will tell us whether the coefficient values are significantly different from 0 or not), and the f and probability of the f value. The f value will tell us whether adding in the independent variables (factors from classes that are not the cognitive decline class) improved the model's overall fit compared to a simpler model.

```python
import statsmodels.formula.api as smf

# formula with interaction variable between cognitive decline factors
formula = (":".join(combined_df.columns[-4:-2]) + ' ~ ' 
           + combined_df.columns[0] + ' + ' 
           + ' + '.join(combined_df.columns[3:-4]))

# run ordinary least squares regression
ols_model_interaction = smf.ols(data=combined_df, formula=formula).fit()

display(ols_model_interaction.summary())
print("F-value: ", ols_model_interaction.fvalue)
print("Probability of F-value: ", ols_model_interaction.f_pvalue)
```

From the summary that was returned by the ols model run on the data, we observe an F-statistic of 1.880 and a probability of the F-statistic occurring to be 0.001242. Because the probability of the f-statistic occurring is very small, particularly smaller than 0.05, that means adding in the factors from the other classes of healthy aging did have a statistically signinficant effect on the model's overall fit compared to a simpler model. This means that there exists some factors that can effectively predict the factors from the cognitive decline class. To examine what these factors are, we will print the p values associated with each coefficient.

```python
# print all pvalues that are less than the significance level of 0.05
def print_consider_pvalues(model):
    for name in model.pvalues.index:
        p_vals = model.pvalues
        if p_vals[name] < 0.05 and (name != 'Intercept' and name != 'YearStart'):
            print(name, ": ", p_vals[name])
    print()

print_consider_pvalues(ols_model_interaction)
```

From the output above, we see that there are seven factors, whose coefficients are significantly different from 0. To better understand what significantly different from 0 means, lets look at the opposite statement. When a coefficient value is not significantly different from 0 or equal to zero, it essentially means adding in the particular factor did not make an impact in improving the model's overall fit. No difference would have been observed if one were to remove the factor from the model. Note that the factor diabetes screening within past 3 years has the smallest p value out of all the factors. The ols regression model seems to indicate that this factor is the most important predicting factor according to its assumptions.

### Random Forest Regressor Model
We will now use a Random Forest Regressor to model our data see if we can draw any similarities or other insights. But before we move on, lets first understand what Random Forests are and why we are using a Random Forest Regressor. Random Forests is a machine algorithm used for classification and regression tasks. In a Random Forest algorithm, a prediction is made for each tree in a collection of decision trees. The predictions are then aggregated to produce a final prediction. One reason for why we want to use a Random Forest algorithm is that it provides a measure of feature importance meaning, we are able to quantify the amount of impact a variable had on a prediction. Moreover, Random Forest algorithms do what is called bootstrap aggregation, which a procedure that involves decision trees being trained on random subsets of data. This is helpful because overfitting and variance is reduced.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

np.random.seed(1)

# creates a bar plot of the top ten most important features
def create_top10_bar(X, y, title):
    
    # split given data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

    # train Random Forest Regressor model
    rf_regressor = RandomForestRegressor()
    rf_regressor.fit(X_train, y_train)

    # rank feature importances where the most important feature is ranked first
    feature_imps = [(name, feat_imp) for (name, feat_imp) in zip(X_cols.columns, rf_regressor.feature_importances_)]
    feature_imps.sort(key=lambda x: x[1], reverse=True)
    
    # only select the top ten features
    feature_imps_plt = feature_imps[0:10]

    feat_names = [name for (name,_) in feature_imps_plt]
    imps = [imp for (_,imp) in feature_imps_plt]

    # make a bar plot with the top 10 most important features
    plt.figure(figsize=(5,5))
    sb.barplot(x=imps, y=feat_names)
    plt.yticks(fontsize=16)
    plt.title("Top 10 features impacting " + str(title))
    plt.xlabel('feature importance')
    plt.show()
    
    return feature_imps
```

The other observation is that oral health tooth retention seems to have less of an impact on predicting the mean death rate of Alzheimer's disease as the percentage of older adults who report having lost 5 or fewer teeth due to decay or gum disease increases (when comparing by state). In consideration of the second observation, please understand the effect of missing values. When we were preparing our data for analysis, we had to impute some missing values with KNN imputation and Mean imputation. Mean imputation can introduce bias as it is prone to outliers. KNN imputation can also introduce bias because it assumes that data is distributed evenly when in reality that might not be the case.

<center><h1>Insights and Conclusion</h1></center>

From the analyses conducted in this tutorial, we discerned that oral health tooth retentiion or more specifically the effect of older people losing 5 or fewer teeth due to gum disease and tooth decay has the strongest potential for determining whether an older person is at an increased risk of developing Alzheimer's disease on the basis of subjective cognitive decline which is the earliest noticeable symptom of dementia and Alzheimer's disease. The interpretation of this information is especially important because depending on the context, people can obtain potentially misleading takeaways. In our case, we saw negative correlations between the percentage of older adults who report having lost 5 or fewer teeth due to decay or gum disease and the factors related to subjective cognitive decline in the Exploratory Data Analysis stage, and a negative coefficient (that was statistically significant) in the ols interaction model in the Analysis and Hypothesis Testing stage of the data science pipeline. From which, one might assume that as the percentage of older adults who have lost 5 or fewer teeth due to gum disease increases then the factors associated with subjective decline worsening decreases so good oral health must increase the risk of developing Alzheimer's Disease right? Wrong! You cannot assume this! Remember, correlation does not equal causation!!! The context in which this information is understood must be considered. For instance, the observations could also mean that fewer teeth loss is seen to be associated with decreased risk of cognitive decline worsening, or that there exists common risk factors between gum disease, tooth decay and Alzheimer's disease. Therefore, more research should be conducted into the relationships between oral health and congitive decline and Alzheimer's disease before coming to any decisive conclusions. What we did in this tutorial was identify that the oral health is most likely to have some sort of underlying connection to subjective cognitive decline that should be investigated. If investigated, this information can be quite valuable due to the early detection nature subjective cognitive decline has in regards to Alzheimer's disease. Steps taken to ensure good oral health might delay or slow the progress of subjective congnitve decline and ultimately the development of Alzheimer's disease.
